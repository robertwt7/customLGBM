```python
import shap
import lightgbm as lgb
import numpy as np
import sklearn
import pandas as pd
from sklearn.datasets import load_svmlight_file
from sklearn.metrics import mean_squared_error

dataPath = '/research/remote/collections/learning_to_rank/LTRv2/'

#Count the amount of queries 
def group_counts(arr):
    d = np.ones(arr.size, dtype=int)
    d[1:] = (arr[:-1] != arr[1:]).astype(int)
    return np.diff(np.where(np.append(d, 1))[0])

dataPaths = [dataPath + 'set1.train.txt', dataPath + 'set1.valid.txt', dataPath + 'set1.test.txt']

def load_lgb_data(dataPaths):
    datas = []
    for data in dataPaths:
        X, y, qid = load_svmlight_file(data, query_id=True)
        datas.append(lgb.Dataset(X, label=y, group=group_counts(qid)))
    return datas


train_data, valid_data, test_data = load_lgb_data(dataPaths)

#Reload the test set to get X and Y to predict
X_test, y_test, qid = load_svmlight_file(dataPath + 'set1.test.txt', query_id=True)
```

```python
bestModel = lgb.Booster(model_file='modelsP05LR/63.1500.0.9.Leaves.tree.Featfrac.txt')
explainer = shap.TreeExplainer(bestModel)
shap_values = explainer.shap_values(X_test)
```

```python
shap.summary_plot(shap_values, X_test)
```

```python
bpath='/research/remote/petabyte/users/robert/Utilities/ltr-baseline/yahoo1/feat_importance/'

top20 = pd.DataFrame()
datas = ["jf.lmart.1500.63.0.05.0.6", "jf.lmart.triskFARO.SOTA", "jf.lmart.triskSARO.SOTA", "jf.lmart.URISK.SOTA"]
name = ["JFNdcg", "TriskFaro", "TriskSaro", "Urisk"]

for data in range(4):
    jfndcg = pd.read_csv(bpath + datas[data], sep=' ', header=None)
    counted = jfndcg.apply(pd.Series.value_counts)

    featureImportance = pd.DataFrame(counted.sum(axis = 1, skipna=True))

    featureImportance = featureImportance.applymap(int)
    featureImportance = featureImportance.sort_values([0], ascending=0)
    temp = featureImportance[:10]
    top20[name[data]] = temp.index
    temp = temp.reset_index(drop=True)
    top20['Feat Value ' + name[data]] = temp[[0]]
    
top20
```

```python
bpath='/research/remote/petabyte/users/robert/Utilities/ltr-baseline/mslr10k/feat_importance/'

top20 = pd.DataFrame()
featureImportance = pd.DataFrame()
datas = ["JFSotaFeatures", "FAROFeatures", "SAROFeatures", "URISKFeatures"]
name = ["JFNdcg", "TriskFaro", "TriskSaro", "Urisk"]

for data in range(4):
    for i in range(5):
        jfndcg = pd.read_csv(bpath + datas[data] + "/fold%d" % (i+1), sep=' ', header=None)
        counted = jfndcg.apply(pd.Series.value_counts)

        featureImportance[i] = counted.sum(axis = 1, skipna=True)
    featureImportancesum = pd.DataFrame(featureImportance.sum(axis = 1, skipna=True) /5)
    featureImportancesum = featureImportancesum.sort_values([0], ascending=0)
    featureImportancesum = featureImportancesum.applymap(int)

    temp = featureImportancesum[:10]
    top20[name[data]] = temp.index
    temp = temp.reset_index(drop=True)
    top20['SplitValue ' + name[data]] = temp[[0]]

top20
```

```python
sns.set()
plt.rcParams['figure.figsize'] = [15,10]
fig = plt.figure()
ax = fig.add_subplot(111)

height = 0.2
ylocs = np.arange(len(top20))
labels = top20['JFNdcg']
values = top20["Feat Value JFNdcg"]
ax.barh(ylocs, values, align='center', height=height)

#To write the amount of occurence in each of the bar
for x, y in zip(values, ylocs):
    ax.text(x + 1, y,
            x,
            va='center')

ax.set_yticks(ylocs)
ax.set_yticklabels(labels)
ax.set_title("Top 20 Features Urisk")
ax.set_xlabel("Feature Importance")
ax.set_ylabel("Features")

plt.show()
```

## Getting MSLR10k 5 fold feature from LGBM (Average between 5 fold)

```python
import lightgbm as lgb

importance = np.zeros((1,137))
for i in range(1,6):
    bestModel = lgb.Booster(model_file='/research/remote/petabyte/users/robert/Utilities/ltr-baseline/mslr10k/model/model.lgbm.lrank.1500.63.0.05.0.4.fold%d' % (i))
    importance = importance + bestModel.feature_importance()
importance = importance/5
newdf = pd.DataFrame()
newdf['value'] = importance[0]
newdf['index'] = newdf.index
newdf.fillna(0)
newdf.applymap(int)
newdf = newdf.sort_values(['value'], ascending = 0).reset_index(drop=True)
newdf
#lgb.plot_importance(bestModel, title='Feature importance', xlabel='Feature importance', 
#                    ylabel='Features',max_num_features=50, figsize=(15,10))
```

```python
def drawplot(df, numbers):
    sns.set()
    plt.rcParams['figure.figsize'] = [15,10]
    fig = plt.figure()
    ax = fig.add_subplot(111)

    topnumbers = df[:numbers]
    height = 0.2
    ylocs = np.arange(len(topnumbers))
    labels = topnumbers['index']
    values = topnumbers["value"]
    ax.barh(ylocs, values, align='center', height=height)

    #To write the amount of occurence in each of the bar
    for x, y in zip(values, ylocs):
        ax.text(x + 1, y,
                x,
                va='center')

    ax.set_yticks(ylocs)
    ax.set_yticklabels(labels)
    ax.set_title("Top %d Features" % (numbers))
    ax.set_xlabel("Feature Importance")
    ax.set_ylabel("Features")

    plt.show()
```

```python
drawplot(newdf, 50)
```
